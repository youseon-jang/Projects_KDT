### 파라미터 중 가장 높은 newton-cg만 사용함!!!

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def get_clf_eval(y_ts, preds=None, pred_proba=None):
    confusion = confusion_matrix(y_ts, preds)
    accuracy = accuracy_score(y_ts, preds)
    precision = precision_score(y_ts, preds)
    recall = recall_score(y_ts, preds)
    f1 = f1_score(y_ts, preds)
    roc_auc = roc_auc_score(y_ts, pred_proba)
    print('오차행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, f1: {3:.4f}, auc: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
solvers = ['newton-cg']
for solver in solvers:
    rf_clf1 = LogisticRegression(solver=solver, max_iter=600)
    rf_clf1.fit(X_train, y_train)
    pred = rf_clf1.predict(X_test)
    pred_prob = rf_clf1.predict_proba(X_test)[:,1]
    print(get_clf_eval(y_test, pred, pred_prob))
```

```python
# 정확도: 0.7301, 정밀도: 0.8333, 재현율: 0.5625, f1: 0.6716, auc: 0.8133
```

```python
rf_clf1 = LogisticRegression()
rf_clf1.fit(X_train, y_train)
pred = rf_clf1.predict(X_test)
pred_prob = rf_clf1.predict_proba(X_test)[:,1]
get_clf_eval(y_test, pred, pred_prob)
```

## LightGBM

```python
X_train = train_data.iloc[:, :-1]
y_train = train_data.iloc[:, -1]

X_test = val_test_data.iloc[:, :-1]
y_test = val_test_data.iloc[:, -1]

X_val, X_ts, y_val, y_ts = train_test_split(X_test, y_test, test_size=.3, random_state=121)

lgbm_wrapper = LGBMClassifier(n_estimators=300, learning_rate=0.015)

evals = [(X_train, y_train),(X_val, y_val)]

lgbm_wrapper.fit(X_train, y_train, early_stopping_rounds=50, eval_metric='logloss', eval_set=evals, verbose=True)

preds = lgbm_wrapper.predict(X_ts)
pred_proba = lgbm_wrapper.predict_proba(X_ts)[:,1]
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def get_clf_eval(y_ts, preds=None, pred_proba=None):
    confusion = confusion_matrix(y_ts, preds)
    accuracy = accuracy_score(y_ts, preds)
    precision = precision_score(y_ts, preds)
    recall = recall_score(y_ts, preds)
    f1 = f1_score(y_ts, preds)
    roc_auc = roc_auc_score(y_ts, pred_proba)
    print('오차행렬')
    print(confusion)
    print('정확도: {0:.4f}, 정밀도: {1:.4f}, 재현율: {2:.4f}, f1: {3:.4f}, auc: {4:.4f}'.format(accuracy, precision, recall, f1, roc_auc))
```

```python
get_clf_eval(y_ts, preds, pred_proba)
```

# Logistic Regression 2

## Data Load

```python
import pandas as pd
import numpy as np
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split
```

```python
train_data = pd.read_csv('./train_features.csv', index_col='점포코드')
val_test_data  = pd.read_csv('./val_test_features.csv', index_col='점포코드')
```

```python
train_data.fillna(0, inplace=True)
val_test_data.fillna(0, inplace=True)
```

```python
train_data.isna().sum()
val_test_data.isna().sum()
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # 평균 0, 분산 1

train_scaled = scaler.fit_transform(train_data.iloc[:, :-1])
val_test_scaled = scaler.fit_transform(val_test_data.iloc[:, :-1])

train_label = train_data.iloc[:, -1]
val_test_label = val_test_data.iloc[:, -1]

X_val, X_ts, y_val, y_ts = train_test_split(val_test_scaled, val_test_label, test_size=.3, random_state=121)
```

```python
from sklearn.metrics import accuracy_score, roc_auc_score

lr_clf = LogisticRegression()
lr_clf.fit(train_scaled, train_label)
lr_preds = lr_clf.predict(X_val)

print('accuracy: {0:.3f}, roc_auc: {1:.3f}'.format(accuracy_score(y_val, lr_preds), roc_auc_score(y_val, lr_preds)))
```

```python
solvers = ['lbfgs', 'liblinear', 'newton-cg', 'sag', 'saga']

for solver in solvers:
    lr_clf = LogisticRegression(solver=solver, max_iter=500)
    lr_clf.fit(train_scaled, train_label)
    lr_preds = lr_clf.predict(X_val)
    
    print('solver:{0}, accuracy:{1:.3f}, roc_auc:{2:.3f}'.format(solver, accuracy_score(y_val, lr_preds), roc_auc_score(y_val, lr_preds)))
```

```python
for solver in solvers:
    lr_clf = LogisticRegression(solver=solver, max_iter=500)
    lr_clf.fit(val_test_scaled, val_test_label)
    lr_preds = lr_clf.predict(X_ts)\
      
    print('solver:{0}, accuracy:{1:.3f}, roc_auc:{2:.3f}'.format(solver, accuracy_score(y_ts, lr_preds), roc_auc_score(y_ts, lr_preds)))
```

```python
from sklearn.model_selection import GridSearchCV

params = {'solver' : ['lbfgs', 'liblinear'], 'penalty' : ['l2', 'l1'], 'C' : [0.01, 0.1, 1, 1, 5, 10]}

lr_clf = LogisticRegression()

grid_clf = GridSearchCV(lr_clf, param_grid=params, scoring='accuracy', cv=3)
grid_clf.fit(val_test_scaled, val_test_label)
print('최적 하이퍼 파라미터:{0}, 최적 평균 정확도:{1:.3f}'.format(grid_clf.best_params_, grid_clf.best_score_))
```
